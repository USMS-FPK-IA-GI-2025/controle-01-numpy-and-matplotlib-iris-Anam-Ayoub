# **Partie A — Importation du Dataset, Arrays, Premières Manipulations**

### **A1. Importation du dataset**
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
# Votre code ici
#=========== BLOC 1 ===========

# Pour importer NumPy
import numpy as np

# Pour importer Matplotlib
import matplotlib.pyplot as plt

# Stockage des données
data = np.genfromtxt('Iris.csv', delimiter=',', dtype=str, skip_header=1)

# Affichage de 'shape'
np.shape(data)
```

### **A2. Séparation des features et des espèces**
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
# Votre code ici
#=========== BLOC 2 ===========

# L'extraction des 4 premières colonnes dans features
# Nous avons commencé par la colonne 1, car la colonne 0 ne contient que les ID
# astype(float) pour transformer les données en flottant
features = data[: , 1:5].astype(float)

# L'extraction de la dernière colonne dans species
species = data[: , -1]
```

### **A3. Indexing**
Affichez les 5 premières lignes de `features` en utilisant l’indexing.


```python
# Votre code ici
#=========== BLOC 3 ===========

print(features[:5 , :])
# :5 indique que nous voulons les 5 premières lignes (indices 0 à 4)
# : indique que nous voulons toutes les colonnes
```

### **A4. Extraction d’une colonne**
Créez un array `sepal_length` contenant uniquement la première colonne de `features`.  
Affichez sa `size` et son `dtype`.


```python
# Votre code ici
#=========== BLOC 4 ===========

sepal_length = features[: , 0]
# : indique que nous voulons toutes les lignes
# 0 s'est l'indice de la première colonne

print(f"Size: {np.size(sepal_length)}\ndtype: {sepal_length.dtype}")
```

### **A5. Slicing par espèce**
Utilisez le slicing pour extraire les 50 premières lignes correspondant à l’espèce *Setosa*.  
Stockez-les dans `setosa_features`.


```python
# Votre code ici
#=========== BLOC 5 ===========

setosa_features = features[:50 , :]
# :50 pour extraire les 50 premières lignes (indices 0 à 49)
```

# **Partie B — Statistiques Basiques et Opérations Mathématiques**

### **B6. Moyenne par colonne**
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
# Votre code ici
#=========== BLOC 6 ===========

moy_col_features = np.mean(features, axis=0)

print(moy_col_features)
```

### **B7. Min et Max**
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
# Votre code ici
#=========== BLOC 7 ===========

min_petal_length = np.min(features[:,2])
max_petal_length = np.max(features[:,2])
# la longueur des pétales se trouve dans la troisième colonne (indice 2)

print(f"Valeur minimale: {min_petal_length}\nValeur maximale: {max_petal_length}")
```

### **B8. Normalisation**
Normalisez les longueurs des pétales en les divisant par leur maximum.
Stockez le résultat dans `normalized_petals`.


```python
# Votre code ici
#=========== BLOC 8 ===========

normalized_petals = np.divide(features[: , 2] , max_petal_length)

print(normalized_petals[:5])
```

### **B9. Somme cumulée (Versicolor)**
Calculez la somme cumulée des largeurs des pétales pour Versicolor.
Affichez les 10 premières valeurs.


```python
# Votre code ici
#=========== BLOC 9 ===========

somme_Versicolor = np.cumsum(features[50:100 , 3])
# [50:100] est utilisé car la classe Versicolor se trouve entre les indices 50 et 99
# (soit de la 51ème à la 100ème ligne)

print(somme_Versicolor[:10])

```

### **B10. Somme cumulée (Verginica)**
Calculez la somme cumulée des longeurs des pétales pour Virginica.
Affichez les 10 premières valeurs.


```python
# Votre code ici
#=========== BLOC 10 ===========

somme_Verginica = np.cumsum(features[100: , 2])
# [100:] est utilisé car la classe Virginica se trouve à partir de l'indice 100
# (soit de la 101ème ligne à la fin)

print(somme_Verginica[:10])
```

# **Partie C — Transformations Avancées avec NumPy**

### **C11. Espèces uniques**
Listez les espèces uniques avec `np.unique` et comptez leurs occurrences.


```python
# Votre code ici
#=========== BLOC 11 ===========

unique_species, counts = np.unique(species, return_counts=True)
# np.unique extrait les valeurs uniques de 'species' et les stocke dans 'unique_species'
# return_counts=True permet de stocker le nombre d'occurrences de chaque espèce dans 'counts'

print("Espèces uniques et leurs occurrences:")

for s,c in zip(unique_species, counts):
  print(f"{s} --> {c}")

```

### **C12. Reshape par espèce**
Transformez `features` en un array de forme `(3, 50, 4)` nommé `features_by_species`.


```python
# Votre code ici
#=========== BLOC 12 ===========

features_by_species = features.reshape(3, 50, 4)
# Redimensionne 'features' en un tenseur de 3 dimensions
# 3 espèces, 50 échantillons(50 lignes), 4 caractéristiques(4 colonnes)

print(f"class1 :\n{features_by_species[0 , :3 , :]}\n")
print(f"class2 :\n{features_by_species[1 , :3 , :]}\n")
print(f"class3 :\n{features_by_species[2 , :3 , :]}")
```

### **C13. Concaténation**
Concaténez horizontalement les moyennes par colonne avec un array de zéros.


```python
# Votre code ici
#=========== BLOC 13 ===========

cat = np.concatenate((moy_col_features , np.zeros(4)))
print(f"Colonne concaténée:\n{cat}")
```

### **C14. Data cleaning (simulation)**
Simulez des valeurs NaN et remplacez-les avec `np.nan_to_num`.


```python
# Votre code ici
#=========== BLOC 14 ===========

# Créons une copie de 'features' pour simuler des NaNs sans altérer l'original
copy_features = features.copy()

# Insérons quelques NaNs à des positions spécifiques pour la simulation
copy_features[0, 0] = np.nan # Première ligne, première colonne
copy_features[5, 2] = np.nan # Sixième ligne, troisième colonne
copy_features[10:12, 1] = np.nan # Lignes 10 et 11, deuxième colonne

print("Array avec NaNs simulés (premières 12 lignes):")
print(copy_features[:12])

cleaned_features = np.nan_to_num(copy_features)
# NaN est remplacé par la valeur par defaut (0)

print("\nArray avec NaNs remplacés (premières 12 lignes):")
print(cleaned_features[:12])
```

### **C15. Broadcasting**
Ajoutez 1 à toutes les valeurs de la première colonne de `features`.


```python
# Votre code ici
#=========== BLOC 15 ===========

sl_p1_features = np.add(features[: , 0], 1)

print(f"Avant: {features[:5 , 0]}")
print(f"Apres: {sl_p1_features[:5]}")
```

### **C16. np.where**
Créez un array binaire indiquant si la longueur des pétales est supérieure à 4.0.


```python
# Votre code ici
#=========== BLOC 16 ===========

whereArr = np.where(features[: , 2] > 4.0 , 1, 0)
# Si la condition est correcte 1 sinon 0

print(whereArr)
```

### **D20. Histogram**
Histogramme de la largeur des sépales (10 bins).


```python
# Votre code ici
#=========== BLOC 20 ===========

plt.hist(features[: , 1], bins=10, color='green', edgecolor='black')
plt.title('Distribution de la largeur des sépales (cm)')
plt.xlabel('Taille (cm)')
plt.ylabel('Fréquence')
plt.show()
```

### **D21. Boxplot par espèce**
Boxplot des longueurs des pétales par espèce.


```python
# Votre code ici
#=========== BLOC 21 ===========

# Extraire la longueur des pétales (index de colonne 2) pour chaque espèce
setosa_petal_length = features[0:50 , 2]
versicolor_petal_length = features[50:100 , 2]
virginica_petal_length = features[100: , 2]

# Créer une liste de ces tableaux pour le boxplot
data_to_plot = [setosa_petal_length, versicolor_petal_length, virginica_petal_length]

plt.boxplot(data_to_plot)

# Définir les étiquettes de l'axe X pour chaque espèce
plt.xticks([1, 2, 3], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])

plt.title('Distribution des longueurs des pétales par espèce')
plt.xlabel('Espèce')
plt.ylabel('Longueur des pétales (cm)')
plt.show()
```

### **E22. Subplots comparatifs**
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python
# Votre code ici
#=========== BLOC 22 ===========

plt.figure(figsize=(10, 3))
plt.subplot(1,2,1)
plt.scatter(features[:50,2], features[:50, 3], color='red')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('Setosa')

plt.subplot(1,2,2)
plt.scatter(features[50:100,2], features[50:100, 3], color='green')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('Versicolor')

plt.suptitle('Setosa vs Versicolor')
plt.show()
```

### **E23. Grille 2x2**
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python
# Votre code ici
#=========== BLOC 23 ===========

# Créer une grille de 2x2 sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Histogramme pour Sepal Length (index 0)
axes[0,0].hist(features[:, 0], bins=10, color='purple', edgecolor='black')
axes[0,0].set_title(f'Distribution de Sepal Length')
axes[0,0].set_xlabel('Sepal Length (cm)')
axes[0,0].set_ylabel('Fréquence')

# Histogramme pour Sepal Width (index 1)
axes[0,1].hist(features[:, 1], bins=10, color='skyblue', edgecolor='black')
axes[0,1].set_title(f'Distribution de Sepal Width')
axes[0,1].set_xlabel('Sepal Width (cm)')
axes[0,1].set_ylabel('Fréquence')

# Histogramme pour Petal Length (index 2)
axes[1,0].hist(features[:, 2], bins=10, color='skyblue', edgecolor='black')
axes[1,0].set_title(f'Distribution de Petal Length')
axes[1,0].set_xlabel('Petal Length (cm)')
axes[1,0].set_ylabel('Fréquence')

# Histogramme pour Petal Width (index 3)
axes[1,1].hist(features[:, 3], bins=10, color='purple', edgecolor='black')
axes[1,1].set_title(f'Distribution de Petal Width')
axes[1,1].set_xlabel('Petal Width (cm)')
axes[1,1].set_ylabel('Fréquence')

plt.tight_layout()
plt.suptitle('Histogrammes des quatre caractéristiques', y=1.02, fontsize=16)
plt.show()
```
